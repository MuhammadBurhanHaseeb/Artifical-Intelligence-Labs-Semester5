Read the last two lines of any text file


```python
# add code here
file_path = "./textfile.txt"  # Replace with the actual path to your text file

try:
    # Code block to work with the file
    # The file is automatically closed when the block is exited
    # Resource management is handled automatically
    with open(file_path, "r") as file:
        # Read all lines from the file
        lines = file.readlines()

        # Display the last two lines
        if len(lines) >= 2:
            print("Last two lines of the file:")
            print(lines[-2].strip())
            print(lines[-1].strip())
        elif len(lines) == 1:
            print("Only one line in the file:")
            print(lines[-1].strip())
        else:
            print("The file is empty.")
except FileNotFoundError:
    print(f"File '{file_path}' not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

Create a dictionary of student **Ali** where the keys are courses and values are total and obtaining marks in each course. Save this dictionary in a text file and numpy file.


```python
# add code here
import numpy as np

# Dictionary for Ali's course marks
ali_marks = {
    "Math": {"total": 100, "obtained": 85},
    "Physics": {"total": 100, "obtained": 75},
    "Chemistry": {"total": 100, "obtained": 90},
    "Computer Science": {"total": 100, "obtained": 92},
}

# Save to a text file
text_file_path = "ali_marks.txt"
with open(text_file_path, "w") as text_file:
    for course, marks in ali_marks.items():
        line = f"{course}: Total = {marks['total']}, Obtained = {marks['obtained']}\n"
        text_file.write(line)

# Save to a NumPy file
np_file_path = "ali_marks.npy"
np.save(np_file_path, ali_marks)
```

Reproduce this figure. **(4 marks)**


```python

# add code here
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.arange(0, 4 * np.pi - 1, 0.1)
sin_curve = np.sin(x)
cos_curve = np.cos(x)

# Plotting the sine and cosine curves
plt.plot(x, sin_curve, label='sin(x)')
plt.plot(x, cos_curve, label='cos(x)')

# Adding labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Curves')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

Reproduce this figure.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate noisy data
noise = np.random.uniform(-20, 20, 100)
x = np.linspace(0, 100, 100) + noise
y = np.linspace(0, 50, 100) + noise

# Scatter plot of the noisy data
plt.scatter(x, y, label='Noisy Data')

# Fit a linear regression line using polyfit
b, m = np.polyfit(x, y, 1)

# Generate predicted values using the linear regression coefficients
y_pred = m * x + b

# Plot the linear regression line
plt.plot(x, y_pred, color='red', label='Linear Regression')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression on Noisy Data')
plt.legend()

# Show the plot
plt.show()
```
