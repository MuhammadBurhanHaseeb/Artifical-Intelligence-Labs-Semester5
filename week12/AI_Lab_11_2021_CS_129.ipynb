Create a dictionary of student **Ali** where the keys are courses and values are total and obtaining marks in each course. Save this dictionary in a text file and numpy file.

```python
import numpy as np

ali = {
    'Chemistry': {'t': 100, 'o': 75},
    'Biology': {'t': 100, 'o': 75},
    'Physics': {'t': 100, 'o': 87},
    'English': {'t': 100, 'o': 9}
}

#NumPy file
numpy_file_path = 'ali.npy'
np.save(numpy_file_path, ali)

#Text file
text_file_path = 'ali.txt'
with open(text_file_path, 'w') as file:
    for course, marks in ali.items():
        file.write(f"{course}: Total Marks - {marks['t']}, Obtained Marks - {marks['o']}\n")
```

Reproduce this figure.  **(4 marks)**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 4 * np.pi - 1, 0.1)
sin_curve = np.sin(x)
cos_curve = np.cos(x)

#code

plt.figure(figsize=(6, 4))
plt.plot(x, sin_curve, label='sin(x)', color='blue', linewidth=2.5, alpha=0.6)
plt.plot(x, cos_curve, label='cos(x)', color='green', linewidth=2.5, alpha=0.6)

plt.ylim(-1, 1)
plt.yticks(np.arange(-1, 1.25, 0.25))

plt.xlim(0, 12)
plt.xticks(np.arange(0, 13, 2))

plt.title('Plot of sin and cos from 0 to 4')

plt.legend(loc='upper right')

plt.xlabel('x values from 0 to 4pi')
plt.ylabel('sin(x) and cos(x)')

plt.grid(True)

plt.show()
```

Reproduce this figure.  

```python
from numpy.polynomial.polynomial import polyfit
import matplotlib.pyplot as plt
import numpy as np

noise = np.random.uniform(-20, 20, 100)
x = np.linspace(0, 100, 100)+noise
y = np.linspace(0, 50, 100)+noise

plt.scatter(
    x, y, marker='o', c='b', edgecolor='r'
)
b, m = polyfit(x, y, 1)
y_pred = m*x+b
plt.plot(
    x, y_pred, '-', c='green'
)
plt.show()
```
